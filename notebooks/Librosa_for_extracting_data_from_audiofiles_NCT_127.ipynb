```python
# NCT 127 Audio Feature Extractor - Colab Version
# Upload individual audio files, extract features one-by-one, and combine

import librosa
import numpy as np
import pandas as pd
import os
from google.colab import files

# === STEP 1: Upload one or more audio files manually ===
print("üìÇ Please upload your .mp3 or .wav files")
uploaded = files.upload()

# === STEP 2: Feature extraction function ===
def extract_features(file_path):
    try:
        y, sr = librosa.load(file_path, sr=22050)

        tempo, _ = librosa.beat.beat_track(y=y, sr=sr)
        centroid = librosa.feature.spectral_centroid(y=y, sr=sr)
        flatness = librosa.feature.spectral_flatness(y=y)
        contrast = librosa.feature.spectral_contrast(y=y, sr=sr)
        mfccs = librosa.feature.mfcc(y=y, sr=sr)
        onset_env = librosa.onset.onset_strength(y=y, sr=sr)
        pitches, mags = librosa.piptrack(y=y, sr=sr)

        pitch_values = pitches[mags > np.median(mags)]
        pitch_range = np.max(pitch_values) - np.min(pitch_values) if len(pitch_values) > 0 else 0

        return {
            'tempo': tempo,
            'mfcc_var': np.var(mfccs),
            'spectral_centroid_mean': np.mean(centroid),
            'spectral_flatness_mean': np.mean(flatness),
            'spectral_contrast_mean': np.mean(contrast),
            'onset_strength_mean': np.mean(onset_env),
            'pitch_range': pitch_range
        }
    except Exception as e:
        print(f"‚ùå Error processing {file_path}: {e}")
        return None

# === STEP 3: Process each uploaded file ===
all_features = []

for file_name in uploaded.keys():
    print(f"\nüîç Extracting from: {file_name}")
    features = extract_features(file_name)
    if features:
        features['track'] = file_name
        all_features.append(features)
        # Save individual feature file
        pd.DataFrame([features]).to_csv(f"{file_name}_features.csv", index=False)
        print(f"‚úÖ Features saved: {file_name}_features.csv")
    else:
        print(f"‚ö†Ô∏è Skipped: {file_name}")

# === STEP 4: Combine all features into one CSV ===
if all_features:
    combined_df = pd.DataFrame(all_features)
    combined_df.to_csv("nct127_combined_features.csv", index=False)
    print("\n‚úÖ All features combined and saved as 'nct127_combined_features.csv'")
else:
    print("‚ùå No features extracted.")
```

```python
from math import pi

# Normalize features for radar plot
features = df.drop(columns='track').copy()
features_normalized = (features - features.min()) / (features.max() - features.min())
features_normalized['track'] = df['track']

# Radar chart for one song
track = "Sticker (1).wav"  # Replace with any track name you want
row = features_normalized[features_normalized['track'] == track].drop(columns='track').iloc[0]

labels = row.index
values = row.values.tolist()
values += values[:1]  # close the circle

angles = [n / float(len(labels)) * 2 * pi for n in range(len(labels))]
angles += angles[:1]

plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], labels, color='grey', size=8)
ax.plot(angles, values, linewidth=2, linestyle='solid')
ax.fill(angles, values, alpha=0.3)
plt.title(f"Audio Feature Profile: {track}")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

# Load your extracted feature dataset
df = pd.read_csv("nct127_all_features_combined.csv")

# Convert 'tempo' column to numeric before scaling
df['tempo'] = df['tempo'].str.strip('[]').astype(float) # Convert tempo to numeric

# Normalize numeric features (excluding 'track')
features = df.drop(columns='track')
scaler = MinMaxScaler()
normalized = scaler.fit_transform(features)
df_norm = pd.DataFrame(normalized, columns=features.columns)
df_norm['track'] = df['track']

# Create complexity and hardness scores
df_norm['complexity_score'] = df_norm[['mfcc_var', 'spectral_contrast_mean', 'pitch_range']].mean(axis=1)
df_norm['hardness_score'] = df_norm[['tempo', 'onset_strength_mean', 'spectral_flatness_mean']].mean(axis=1)

# Create summary table
summary_table = df_norm[['track', 'complexity_score', 'hardness_score']].sort_values(by='complexity_score', ascending=False)
summary_table.reset_index(drop=True, inplace=True)
print(summary_table)

# Plot: Complexity vs Hardness
plt.figure(figsize=(10, 6))
sns.scatterplot(data=summary_table, x='complexity_score', y='hardness_score', hue='track', s=100)
plt.title("Complexity vs. Hardness of NCT 127 Songs")
plt.xlabel("Musical Complexity Score")
plt.ylabel("Perceptual Hardness Score")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
